```python
content=[]
for l1 in range(len(area_page)):
    for l2 in range(len(name)):
        for l3 in range(len(page)):
            url2='https://{0}.lianjia.com{1}{2}/{3}'.format('bj',area_page[l1],name[l2],page[l3])
            response2=requests.get(url2,headers)
            soup2= BeautifulSoup(response2.text,'html.parser')
            count2=soup2.find_all('h2',class_="total fl")[0].find('span').text

            for x in range(len(soup2.find(class_='sellListContent').find_all('li'))):
                d=dict()
                if x==0:
                    sss=soup2.find(class_='sellListContent').find('li')
                    ss=soup2.find(class_='sellListContent').find('li').find('div').find_all('div')
                else:
                    sss=sss.next_sibling
                    ss=sss.find('div').find_all('div')
                for y in range(len(ss)):
                    if y==0:
                        s=sss.find('div').find('div')
                        label=s['class'][0]
                    else:
                        try:
                            s=s.next_sibling
                            label=s['class'][0]
                            d[label]=s.get_text()
                        except:
                            pass
                    print('{0}-{1}'.format(y,label))
                content.append(d)
```
